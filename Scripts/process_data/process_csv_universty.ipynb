```python
import pandas as pd

def read_xlsx_from_sheet2(filename):
    # Leer solo la hoja 2 del archivo XLSX
    df = pd.read_excel(filename, sheet_name="Sheet2",header=0)  # Las hojas se indexan desde 0
    
    # Descartar la primera fila y reiniciar el índice
    df.columns = df.iloc[0]
    df = df.iloc[1:].reset_index(drop=True)

    return df

# Uso de la función
filename = '/home/user/Desktop/MateoCodes/WebScrapingLinkedin/documentacion/RE_EGR_ACR_.xlsx'
data = read_xlsx_from_sheet2(filename)
print(data)
```

```python
# Solo quedarse con los datos necesarios
data = data.loc[:,['PERIODO_EGRESO','PERIODO_GRADO','SEDE','FACULTAD','UAB','PROGRAMA','PLAN','NIVEL','NOMBRES','APELLIDO1','APELLIDO2','CONVOCATORIA','CORREO']]
```

```python
# Ahora solo procesar las personas que son de la facultad de ingenieria para los programas de pregrado
data = data[data['FACULTAD'] == 'FACULTAD DE INGENIERÍA']
filtro = data[~data['PROGRAMA'].str.contains('MAESTRÍA|ESPECIALIZACIÓN|DOCTORADO', case=False, na=False)]
```

```python
# Filtro para que solo se tengan encuenta a los egresados desde el 2012 al 2023

filtro_fechas =filtro[filtro['PERIODO_EGRESO'].str.contains('201[2-9]|202[0-9]', na=False)]

filtro_fechas
```
